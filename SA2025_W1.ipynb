```python
# Load the dataset
data = pd.read_csv("Cars.csv")
```

###  7) Print a list of all the unique mpg values





```python
# List of all unique MPG values
unique_mpg = data['mpg'].unique().tolist()
print("Unique MPG values:", unique_mpg)
```

### 8) Create a column which contains the horsepower divided by weightas its metric and make this new column the index.

```python
# Create a new column: horsepower divided by weight
data['horsepower'] = pd.to_numeric(data['horsepower'], errors='coerce')
data['weight'] = pd.to_numeric(data['weight'], errors='coerce')
data['hp_per_weight'] = data['horsepower'] / data['weight']
data[['horsepower', 'weight', 'hp_per_weight']].head()
```

### 1. What is name of car that has the highest horsepower?  

```python
# Car with highest horsepower
data['horsepower'] = pd.to_numeric(data['horsepower'], errors='coerce')
max_hp_car = data['horsepower'].idxmax()
print("Car with highest horsepower:", max_hp_car)
```

### 2. How many cars have mpg  35?

```python
# Count of cars with MPG  35
high_mpg_count = data[data['mpg'] >= 35].shape[0]
print("Cars with MPG  35:", high_mpg_count)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
# Most common origin where horsepower > 100 and weight < 3000
filtered = data[(data['horsepower'] > 100) & (data['weight'] < 3000)]
common_origin = filtered['origin'].mode()[0]
print("Most common origin (hp > 100, weight < 3000):", common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
# Mean acceleration of Japanese cars
mean_acc_japan = data[data['origin'] == 'japan']['acceleration'].mean()
print("Mean acceleration (Japan):", round(mean_acc_japan, 2))
```

### 5. Which year had the highest average mpg?

```python
# Year with highest average MPG
avg_mpg_by_year = data.groupby('model_year')['mpg'].mean()
best_year = avg_mpg_by_year.idxmax()
print("Year with highest average MPG:", best_year)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# Car with best hp-to-weight ratio above median MPG
above_median = data[data['mpg'] > data['mpg'].median()]
max_ratio_car = above_median['hp_per_weight'].idxmax()
print("Best hp/weight car with MPG > median:", max_ratio_car)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
# Multi-line plot of average MPG over years
plt.figure(figsize=(10,6))
sns.lineplot(x='model_year', y='mpg', data=data, estimator='mean')
plt.title("Average MPG Over the Years")
plt.xlabel("Model Year")
plt.ylabel("Average MPG")
plt.grid(True)
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
# Scatter plot: horsepower vs weight, colored by origin
# Ensure these columns are numeric
data['horsepower'] = pd.to_numeric(data['horsepower'], errors='coerce')
data['weight'] = pd.to_numeric(data['weight'], errors='coerce')

# Drop rows with missing values in those columns
scatter_data = data.dropna(subset=['horsepower', 'weight', 'origin'])

plt.figure(figsize=(8,6))
sns.scatterplot(data=scatter_data, x='horsepower', y='weight', hue='origin')
plt.title("Horsepower vs Weight by Origin")
plt.xlabel("Horsepower")
plt.ylabel("Weight")
plt.grid(True)
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# Consistent models: same model in multiple years with small MPG variation
data_reset = data.reset_index()
grouped = data_reset.groupby('name')
consistent_models = [name for name, group in grouped if group['mpg'].std() < 1.0 and group['model_year'].nunique() > 1]
print("Consistent models:", consistent_models)
```
